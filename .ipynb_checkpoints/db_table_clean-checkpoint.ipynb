```python


# Load Skilled Personnel dataset
skilled_data = pd.read_csv("cs3200project/raw-data/Skilled_Personnel_cleaned.csv")  # Update the file name to match your dataset

# Extract unique regions
regions = skilled_data[["ParentLocationCode", "ParentLocation"]].drop_duplicates()
regions.columns = ["region_id", "region_name"]

# Save cleaned regions to a CSV file
regions.to_csv("cleaned_regions.csv", index=False)

# Extract unique countries
countries = skilled_data[["SpatialDimValueCode", "Location", "ParentLocationCode"]].drop_duplicates()
countries.columns = ["geo_code", "country_name", "region_id"]

# Save cleaned countries to a CSV file
countries.to_csv("cleaned_countries.csv", index=False)
```

```python
# Load datasets
maternal_mortality = pd.read_csv("cs3200project/raw-data/Mother_Mortality_Rate_cleaned.csv")
skilled_personnel = pd.read_csv("cs3200project/raw-data/Skilled_Personnel_cleaned.csv")
doctors_per_capita = pd.read_csv("cs3200project/raw-data/Doctors_Per_Capita_cleaned.csv")

# Load country mapping
countries = pd.read_csv("cs3200project/raw-data/cleaned_countries.csv")

# Step 1: Extract Maternal Mortality
maternal_mortality_cleaned = maternal_mortality[["GEO_NAME_SHORT", "RATE_PER_100000_N", "DIM_TIME"]].copy()
maternal_mortality_cleaned.rename(columns={
    "GEO_NAME_SHORT": "country_name", 
    "RATE_PER_100000_N": "value", 
    "DIM_TIME": "data_year"
}, inplace=True)
maternal_mortality_cleaned["indicator_id"] = 1  # Maternal Mortality indicator
maternal_mortality_cleaned["indicator_name"] = "Maternal Mortality Ratio"
maternal_mortality_cleaned = maternal_mortality_cleaned.merge(countries, on="country_name", how="inner")
maternal_mortality_cleaned = maternal_mortality_cleaned[["geo_code", "indicator_id", "indicator_name", "value", "data_year"]]

# Step 2: Extract Skilled Birth Attendance
skilled_personnel_cleaned = skilled_personnel[["Location", "FactValueNumeric", "Period"]].copy()
skilled_personnel_cleaned.rename(columns={
    "Location": "country_name", 
    "FactValueNumeric": "value", 
    "Period": "data_year"
}, inplace=True)
skilled_personnel_cleaned["indicator_id"] = 2  # Skilled Birth Attendance indicator
skilled_personnel_cleaned["indicator_name"] = "Skilled Birth Attendance (%)"
skilled_personnel_cleaned = skilled_personnel_cleaned.merge(countries, on="country_name", how="inner")
skilled_personnel_cleaned = skilled_personnel_cleaned[["geo_code", "indicator_id", "indicator_name", "value", "data_year"]]

# Step 3: Extract Doctors Per Capita
doctors_per_capita_cleaned = doctors_per_capita[["country", "DoctorsPerCapita_DensityOfMedicalDoctors_Per10000", "DoctorsPerCapita_DensityOfMedicalDoctors_DataYear"]].copy()
doctors_per_capita_cleaned.rename(columns={
    "country": "country_name", 
    "DoctorsPerCapita_DensityOfMedicalDoctors_Per10000": "value", 
    "DoctorsPerCapita_DensityOfMedicalDoctors_DataYear": "data_year"
}, inplace=True)
doctors_per_capita_cleaned["indicator_id"] = 3  # Doctors Per Capita indicator
doctors_per_capita_cleaned["indicator_name"] = "Doctors Per 10,000 People"
doctors_per_capita_cleaned = doctors_per_capita_cleaned.merge(countries, on="country_name", how="inner")
doctors_per_capita_cleaned = doctors_per_capita_cleaned[["geo_code", "indicator_id", "indicator_name", "value", "data_year"]]

# Step 4: Extract Nurses Per Capita (Indicator 4)
nurses_per_capita = doctors_per_capita[["country", "DoctorsPerCapita_NursingAndMidwiferyPersonnel_Per10000", "DoctorsPerCapita_NursingAndMidwiferyPersonnel_DataYear"]].copy()
nurses_per_capita.rename(columns={
    "country": "country_name", 
    "DoctorsPerCapita_NursingAndMidwiferyPersonnel_Per10000": "value", 
    "DoctorsPerCapita_NursingAndMidwiferyPersonnel_DataYear": "data_year"
}, inplace=True)
nurses_per_capita["indicator_id"] = 4  # Nurses Per Capita
nurses_per_capita = nurses_per_capita.merge(countries, on="country_name", how="inner")
nurses_per_capita = nurses_per_capita[["geo_code", "indicator_id", "value", "data_year"]]

# Step 5: Extract Dentists Per Capita (Indicator 5)
dentists_per_capita = doctors_per_capita[["country", "DoctorsPerCapita_DentistsPer10000", "DoctorsPerCapita_DentistsPer10000_DataYear"]].copy()
dentists_per_capita.rename(columns={
    "country": "country_name", 
    "DoctorsPerCapita_DentistsPer10000": "value", 
    "DoctorsPerCapita_DentistsPer10000_DataYear": "data_year"
}, inplace=True)
dentists_per_capita["indicator_id"] = 5  # Dentists Per Capita
dentists_per_capita = dentists_per_capita.merge(countries, on="country_name", how="inner")
dentists_per_capita = dentists_per_capita[["geo_code", "indicator_id", "value", "data_year"]]

# Step 6: Extract Pharmacists Per Capita (Indicator 6)
pharmacists_per_capita = doctors_per_capita[["country", "DoctorsPerCapita_PharmacistsPer10000", "DoctorsPerCapita_PharmacistsPer10000_DataYear"]].copy()
pharmacists_per_capita.rename(columns={
    "country": "country_name", 
    "DoctorsPerCapita_PharmacistsPer10000": "value", 
    "DoctorsPerCapita_PharmacistsPer10000_DataYear": "data_year"
}, inplace=True)
pharmacists_per_capita["indicator_id"] = 6  # Pharmacists Per Capita
pharmacists_per_capita = pharmacists_per_capita.merge(countries, on="country_name", how="inner")
pharmacists_per_capita = pharmacists_per_capita[["geo_code", "indicator_id", "value", "data_year"]]


# Combine all indicators into one dataset
country_indicators = pd.concat([maternal_mortality_cleaned, skilled_personnel_cleaned, doctors_per_capita_cleaned, nurses_per_capita, dentists_per_capita, pharmacists_per_capita])

# Save to CSV
country_indicators.to_csv("cleaned_country_indicators.csv", index=False)
```
